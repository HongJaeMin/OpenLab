```python
# PC 쇼핑몰로 구매한 고객 중, 구매수량이 5개 이상인 고객만 추출 
cond1 = (df1['주문경로']=='PC쇼핑몰')
cond2 = (df1['수량']>=5)
df1.loc[cond1 & cond2] #  & : 두 조건을 모두 만족되는 데이터를 추출 
```

```python
cond1 = (df1['주문경로']=='PC쇼핑몰')
cond2 = (df1['주문경로']=='네이버 페이')
df1.loc[cond1 | cond2] # | : 두 조건 중 하나이상 만족되는 데이터를 추출 
```

```python
df1.pivot_table(index='주문경로',values='상품구매금액',aggfunc='median')
```

```python
df1.pivot_table(index='주문경로',values='상품구매금액',
                aggfunc=['mean','median','max'])
# 데이터 개수가 많으면 많을 수록, 이상치에 의한 변동효과가 줄어든다 
```

```python
df1.pivot_table(index='주문경로',values='상품구매금액',
                aggfunc=['mean','median','max']).to_excel('result4.xlsx')
```
